# IRIS-CLASSIFICATION-PROJECT

In this project we get the iris dataset (Classification) from sklearn module 

we have 3 type of iris flowers --> **setosa , versicolor , virginica**

we have features --> <u>sepal length (cm) , sepal width (cm) , petal length (cm) , petal width (cm)</u>

for each **Target** we have **50 smaples** , **total of 150 smaples**

## Exact Problem
We are trying to use the attributes of flower to preedict the species of flower
bye using attribute **speal Length / Width and petal Length / Width** we are trying to predict whether the flower is **setosa , versicolor , virginica**

```python
##Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
```

```python
sns.set()
```

```python
#Loading the dataset
data = datasets.load_iris()
```

```python
#Creating a Dataframe from data
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
df['target_name'] = df['target'].map({0:'Setosa' , 1:'versicolor' , 2:'virginica'})
```

# MODELING

## Train Test Split

```python
x = df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']].values
y = df['target'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3, random_state=42)
```

# ***BASELINE***

What is the Simplest Model here?? 
it is a model that is randomly geussing the species , since there are 150 sample and 50 of each species , even if model geuss same species for every datapoint still the accuracy is 33% --> 0.33 

This is the lowest our model should achive because if lower then 0.33 then its better to geuss same result instead of using ML (lol)

```python
model = LogisticRegression()
model.fit(x_train , y_train)
pre = model.predict(x_test)
print("R2 Score : ",r2_score(y_test,pre))
```
